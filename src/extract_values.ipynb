```python
import re
import matplotlib.pyplot as plt

result = []
episode = []
reward = []
loss = []
accuracy = []
with open('d:/elte/ELTE-IK_2021_2022_2/artificial_intelligence_in_processes_and_automation/theory/project/history.txt') as f:
    index = 0
    for count, line in enumerate(f, start=1):
        if count % 2 == 0:
            match_loss = re.search('loss: (.*) -', line)
            match_accuracy = re.search('accuracy: (\d+\.\d+)', line)
            new.append(float(match_loss.group(1)))
            loss.append(float(match_loss.group(1)))
            new.append(float(match_accuracy.group(1)))
            accuracy.append(float(match_accuracy.group(1)))
            result.append(new)            
            index += 1
        else:
            new = []
            match_episode = re.search('episode (\d+)', line)
            match_reward_plus = re.search('reward was : (\d+\.\d+)', line)
            match_reward_minus = re.search('reward was : (\-\d+\.\d+)', line)
            new.append(int(match_episode.group(1)))
            episode.append(int(match_episode.group(1)))
            if match_reward_minus:
                new.append(float(match_reward_minus.group(1)))
                reward.append(float(match_reward_minus.group(1)))
            else:
                new.append(float(match_reward_plus.group(1)))
                reward.append(float(match_reward_plus.group(1)))
            
            
print(reward)


```

```python
from matplotlib.pyplot import figure

figure(figsize=(15, 6), dpi=100)
font = {'family':'serif','color':'black','size':13}

plt.xlabel("Episode", fontdict = font)
plt.ylabel("Reward", fontdict = font)
plt.plot(episode, reward)
```

```python
reward_average = []
episode_average = []

sum = 0
index = 0
for i in reward:
    index += 1
    sum += i
    if index % 20 == 0:
        episode_average.append(index)
        reward_average.append(sum/20)
        sum = 0

figure(figsize=(15, 6), dpi=100)
font = {'family':'serif','color':'black','size':13}

plt.xlabel("Episode (increments of 20)", fontdict = font)
plt.ylabel("Average reward over every 20 episodes", fontdict = font)
plt.plot(episode_average, reward_average)
```
